```python
df = pd.read_csv('./data/vnindex_historical.csv', usecols=['Date', 'Close'], parse_dates=['Date'])
df.columns = ['ds', 'y']
```

```python
df['y'] = df['y'].str.replace(',', '').astype(float)
df['ds'] = pd.to_datetime(df['ds'], format='%Y/%m/%d')
```

```python
import datetime

df = df[(df['ds'] > "2018-01-01") & (df['ds'] < "2019-07-01")]
df
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(df['ds'],df['y'])

# Set the chart title and axis labels
ax.set_title('Line Chart')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# Display the chart
plt.show()
```
